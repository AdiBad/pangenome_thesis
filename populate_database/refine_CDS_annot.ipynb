# KEEP ONLY CDS SPECIFIC INFO FROM THE HITHERTO CONCATENATED+BASH_PROCESSED GFF FILE CALLED cds_annot.csv

```python
import pandas as pd
import re
```

```python
reqinfo = pd.read_csv('cds_annot.tsv',sep='\t',header=None)
```

```python
reqinfo.columns=['CDS_start','CDS_end','strand','attribute']
```

```python
reqinfo.head()
```

```python
reqinfo[['ID','Other_attributes']] = reqinfo['attribute'].str.split(';',1,expand=True)
```

# HERE ONWARDS JUST CLEAR OUT THE ATTRIBUTE DATA INTO CLEAN SEPARATE COLUMNS

```python
reqinfo.head()
```

```python
reqinfo.tail()
```

```python
reqinfo[['Other_stuff','Name']] = reqinfo['attribute'].str.split('Name=',expand=True)
```

```python
reqinfo['Gene_name'] = reqinfo['Name'].str.split(';').str[0]
```

```python
reqinfo['COG_ID'] = reqinfo['Name'].str.extract('COG:(\w+);')
reqinfo['UNIPROT_ID'] = reqinfo['Name'].str.extract('UniProtKB:(\w+);')
reqinfo['ID'] = reqinfo['ID'].str.extract('ID=(\w+)')
```

```python
reqinfo.head(7)
```

```python
final_df=reqinfo[['ID','Gene_name','COG_ID','UNIPROT_ID','strand','CDS_start','CDS_end']]
```

```python
final_df.head()
```

```python
final_df.to_csv('cds_annot.csv',index=None)
```
