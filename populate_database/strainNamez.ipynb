```python
#PA matrix from panaroo
dat = pd.read_csv('gene_presence_absence.csv')
```

```python
dat.head()
```

```python
df=list(dat.loc[1,:][3:])
```

```python
names=list(dat.columns[3:])
```

```python
referStrain={}
for n in range(len(names)):
    referStrain[df[n].split("_")[0]]=names[n].split(".")[0].split("_")[1]
```

```python
f = open("referStrain.txt","w")
f.write( str(referStrain) )
f.close()
```
