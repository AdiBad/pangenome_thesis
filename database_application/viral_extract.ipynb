```python
row=[]

for file in os.listdir():
    flag = 0
    if file.endswith('summary.txt'):
        
        #Read this file
        sum_text=open(file,'r')
        sum_table=sum_text.readlines()
        
        #Reinitialize the header and append the rest of the table
        for line in enumerate(sum_table):
                     
            #Just to confirm if some file does not start the summary at line 33
            if "----" in line[1]:
                print(file, '-> Split text at ',line[0])
                columns=header.split()
                flag = 1
            if flag == 1:
                if(len(line[1].split())>1):
                    row.append(line[1].split())
            header = line[1]
```

```python
summary_df = pd.DataFrame(row,columns=columns)
summary_df[['Strain', 'Positions']] = summary_df['REGION_POSITION'].str.split(':', 1, expand=True)
summary_df[['CDS_start', 'CDS_stop']] = summary_df['Positions'].str.split('-', 1, expand=True)
summary_df['Strain']=summary_df['Strain'].str[:-2]
summary_df=summary_df.rename(columns={"MOST_COMMON_PHAGE_NAME(hit_genes_count)": "MOST_COMMON_PHAGE_NAME_and_genes_count"})
```
